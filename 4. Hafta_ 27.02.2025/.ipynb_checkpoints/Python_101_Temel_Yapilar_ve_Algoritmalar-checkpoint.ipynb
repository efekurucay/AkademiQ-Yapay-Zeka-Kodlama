# Bölüm 1: Temel Yapılar ve Kullanımları

## 1.1 If-Else Yapısı
**Açıklama:** Koşula bağlı işlem yapılmasını sağlar.

**Sözdizimi:**
```python
if koşul:
    # koşul doğruysa çalışır
else:
    # koşul yanlışsa çalışır
```
---

## 1.2 Match-Case Yapısı
**Açıklama:** Python 3.10+ ile gelen, farklı durumları kontrol etmek için kullanılır.

**Sözdizimi:**
```python
match değer:
    case durum1:
        # durum1 doğruysa çalışır
    case durum2:
        # durum2 doğruysa çalışır
    case _:
        # hiçbir durum eşleşmezse çalışır
```
---

## 1.3 For Döngüsü
**Açıklama:** Belirli aralıktaki işlemleri tekrar etmek için kullanılır.

**Sözdizimi:**
```python
for değişken in aralık:
    # her adımda çalışır
```
---

## 1.4 While Döngüsü
**Açıklama:** Koşul doğru oldukça işlemleri tekrarlar.

**Sözdizimi:**

```python
    # koşul doğruysa çalışır
```


# Bölüm 2: Araştırma Görevi (While Döngüsü)

### While döngüsü nedir ve for döngüsünden farkı nedir?
- **While döngüsü:** Belirsiz tekrar sayısına sahip döngüler için kullanılır.
- **For döngüsü:** Belirli tekrar sayısına sahip döngülerde kullanılır.

### Hangi durumlarda while döngüsü daha kullanışlıdır?
- Tekrar sayısı bilinmeyen durumlarda (örneğin kullanıcı girişleri, tahmin oyunları).

### Örnek:
```python
# Kullanıcı 0 girene kadar sayıların toplamını hesaplar
toplam = 0
while True:
    sayi = int(input("Sayı girin (0 çıkış): "))
    if sayi == 0:
        break
    toplam += sayi
print("Toplam:", toplam)
```


# Bölüm 3: Algoritma Soruları ve Çözümleri

## Örnek Çözüm - Soru 1: İki Sayının Toplamı


````

# Kullanıcıdan iki sayı al ve toplamını yazdır
sayi1 = int(input("Birinci sayıyı girin: "))
sayi2 = int(input("İkinci sayıyı girin: "))
toplam = sayi1 + sayi2
print("Toplam:", toplam)
````

## Örnek Çözüm - Soru 10: Asal Sayı Kontrolü

````

# Asal sayı kontrolü
sayi = int(input("Bir sayı girin: "))
asal = True
if sayi <= 1:
    asal = False
else:
    for i in range(2, int(sayi ** 0.5) + 1):
        if sayi % i == 0:
            asal = False
            break
print("Asal" if asal else "Asal değil")
````

## Örnek Çözüm - Soru 11: Faktöriyel Hesaplama

````

# Faktöriyel hesaplama
sayi = int(input("Bir sayı girin: "))
faktoriyel = 1
i = 1
while i <= sayi:
    faktoriyel *= i
    i += 1
print("Faktöriyel:", faktoriyel)
````

## Örnek Çözüm - Soru 12: Sayı Tahmin Oyunu

````

# Sayı tahmin oyunu
import random
hedef = random.randint(1, 100)
tahmin = 0
while tahmin != hedef:
    tahmin = int(input("Tahmininiz: "))
    if tahmin < hedef:
        print("Daha büyük sayı girin!")
    elif tahmin > hedef:
        print("Daha küçük sayı girin!")
    else:
        print("Doğru tahmin!")
````

## Örnek Çözüm - Soru 31: Not Harfi Belirleme

````

# Not harfi belirleme
notu = int(input("Notunuzu girin: "))
if notu >= 90:
    print("A")
elif notu >= 80:
    print("B")
elif notu >= 70:
    print("C")
elif notu >= 60:
    print("D")
else:
    print("F")
````

## Örnek Çözüm - Soru 33: Artık Yıl Kontrolü

````

# Artık yıl kontrolü
yil = int(input("Yıl giriniz: "))
if (yil % 4 == 0 and yil % 100 != 0) or (yil % 400 == 0):
    print("Artık yıl")
else:
    print("Artık yıl değil")
````
